SELECSTAR 혐오 표현 데이터 전처리

```python
import os
import json
import glob
import pandas as pd
```

```python
path_to_json_files = 'C:/Users/wjdgh/Desktop/Selectstar_Tunip_HUMANE+Lab_Opendataset/Selectstar_Tunip_HUMANE Lab_opendata'
json_file_names = [filename for filename in os.listdir(path_to_json_files) if filename.endswith('.json')]

data = []

for json_file_name in json_file_names:
    json_file_path = os.path.join(path_to_json_files, json_file_name)
    with open(json_file_path, 'r', encoding='UTF-8-SIG') as json_file:
        json_text = json.load(json_file)
        data.append(json_text)
        
df = pd.DataFrame(data)
```

```python
df = df[['문장','욕설']]
df.columns = ['content','label']
```
